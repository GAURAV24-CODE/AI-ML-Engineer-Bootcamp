```python
#Q5. Swap two variables.Method 1 (Pythonic)
a = 10
b = 20

a, b = b, a

print("a =", a)
print("b =", b)
```

```python
#Method 2 (Using Temporary Variable)
a = 10
b = 20

temp = a
a = b
b = temp

print("a =", a)
print("b =", b)
```

```python
#Q8. Calculate Simple Interest. SI = (P × R × T) / 100
P = float(input("Enter Principal: "))
R = float(input("Enter Rate: "))
T = float(input("Enter Time: "))

SI = (P * R * T) / 100

print("Simple Interest =", SI)
```

```python
#Q9. Find whether a number is even or odd.
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
#Q10. Write a program to calculate the average of 5 numbers.
n1 = float(input("Enter number 1: "))
n2 = float(input("Enter number 2: "))
n3 = float(input("Enter number 3: "))
n4 = float(input("Enter number 4: "))
n5 = float(input("Enter number 5: "))

average = (n1 + n2 + n3 + n4 + n5) / 5

print("Average =", average)
```
