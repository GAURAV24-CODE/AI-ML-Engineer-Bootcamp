# Variables

A **variable** is a container used to store data.

Example:

```python
name = "Gaurav"
```


````python
# ==========================================
# Program 2 : Variables
# ==========================================

# Creating variables

name = "Gaurav"
age = 21
city = "Jalgaon"

# Display variables

print("Name :", name)
print("Age  :", age)
print("City :", city)
````

# Data Types

Python supports many data types.

Today's data types:

- int
- float
- str
- bool

````python
# ==========================================
# Program 3 : Data Types
# ==========================================

name = "Python"
age = 21
height = 5.8
is_student = True

print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
````

# Arithmetic Operators

Arithmetic operators perform mathematical calculations.

Operators:

- +
- -
- *
- /
- //
- %
- **

````python
# ==========================================
# Program 4 : Arithmetic Operators
# ==========================================

a = 20
b = 6

print("Addition       :", a + b)
print("Subtraction    :", a - b)
print("Multiplication :", a * b)
print("Division       :", a / b)
print("Floor Division :", a // b)
print("Modulus        :", a % b)
print("Exponent       :", a ** b)
````

# User Input

The **input()** function is used to take input from the user.

By default, input is stored as a string.

````python
# ==========================================
# Program 5 : User Inputgaurav

# ==========================================

name = input("Enter your name: ")

print("Welcome", name)
````

# Type Casting

Type casting converts one data type into another.

Functions:

- int()
- float()
- str()

````python
# ==========================================
# Program 6 : Type Casting
# ==========================================

number = "100"

print(type(number))

number = int(number)

print(type(number))
````
