# 📌 What is Artificial Intelligence (AI)?

Artificial Intelligence (AI) is a branch of computer science that focuses on creating systems capable of performing tasks that normally require human intelligence.

These tasks include:

- 🗣 Speech Recognition
- 👀 Image Recognition
- 🌐 Language Translation
- 🚗 Self-driving Cars
- 💬 Chatbots
- 🎯 Recommendation Systems

---

# 📌 What is Machine Learning (ML)?

Machine Learning is a subset of Artificial Intelligence.

Instead of following fixed rules, machine learning models learn patterns from data and improve their performance through experience.

Examples:

- 📧 Spam Detection
- 💳 Fraud Detection
- 🎬 Movie Recommendations
- 🏥 Disease Prediction

---

# 📌 What is Deep Learning (DL)?

Deep Learning is a specialized branch of Machine Learning that uses artificial neural networks with many layers.

It is especially effective for:

- 🖼 Computer Vision
- 🗣 Speech Recognition
- 🤖 Generative AI
- 🌐 Natural Language Processing

---

# 📊 AI vs ML vs DL

```
Artificial Intelligence
        │
        ▼
Machine Learning
        │
        ▼
Deep Learning
```

| AI | ML | DL |
|----|----|----|
| Broad field | Learns from data | Neural networks |
| Rule-based or learning | Data-driven | Large datasets |
| Decision making | Pattern learning | Complex pattern recognition |
