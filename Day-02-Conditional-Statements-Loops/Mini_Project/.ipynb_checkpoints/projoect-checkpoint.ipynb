```python
#🌟 Bonus Challenge : Menu-Driven Program
print("1. Even / Odd")
print("2. Grade Calculator")
print("3. Multiplication Table")
print("4. Exit")

choice = int(input("Enter your choice: "))

if choice == 1:

    num = int(input("Enter Number: "))

    if num % 2 == 0:
        print("Even Number")
    else:
        print("Odd Number")

elif choice == 2:

    marks = int(input("Enter Marks: "))

    if marks >= 90:
        print("Grade A")
    elif marks >= 75:
        print("Grade B")
    elif marks >= 60:
        print("Grade C")
    else:
        print("Grade D")

elif choice == 3:

    num = int(input("Enter Number: "))

    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")

elif choice == 4:
    print("Program Exited Successfully.")

else:
    print("Invalid Choice.")
```
