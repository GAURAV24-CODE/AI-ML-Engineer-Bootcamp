# 📖 What are Conditional Statements?

A **conditional statement** allows a program to make decisions.

Instead of executing every line of code, Python checks whether a condition is **True** or **False** and then decides what to execute.

---

---

# 🔹 if Statement

The **if** statement executes a block of code only when the condition is **True**.

### Syntax

```python
if condition:
    statement
```

````python
# ==========================================================
# Program 1 : if Statement
# ==========================================================

# Take age as input from the user
age = int(input("Enter your age: "))

# Check if the user is eligible to vote
if age >= 18:
    print("✅ You are eligible to vote.")
````

# 🔹 if...else Statement

The **if...else** statement is used when there are **two possible outcomes**.

- If the condition is **True**, the `if` block executes.
- Otherwise, the `else` block executes.

---

## 📌 Syntax

```python
if condition:
    # Code if condition is True
else:
    # Code if condition is False
```


````python
# ==========================================================
# Program 2 : if...else Statement
# ==========================================================

# Take age as input from the user
age = int(input("Enter your age: "))

# Check voting eligibility
if age >= 18:
    print("✅ You are eligible to vote.")
else:
    print("❌ You are not eligible to vote.")
````

# 🔹 if...elif...else Statement

Use **if...elif...else** when there are **multiple conditions**.

Python checks each condition from top to bottom.
Once a condition becomes **True**, the remaining conditions are skipped.

---

## 📌 Syntax

```python
if condition1:
    # Code
elif condition2:
    # Code
else:
    # Code
```

````python
# ==========================================================
# Program 3 : if...elif...else Statement
# ==========================================================

# Take marks as input
marks = int(input("Enter your marks: "))

# Find the grade
if marks >= 90:
    print("🏆 Grade A")
elif marks >= 75:
    print("🥈 Grade B")
elif marks >= 60:
    print("🥉 Grade C")
else:
    print("📘 Grade D")
````

# 🔹 Nested if Statement

A **Nested if** means placing one `if` statement inside another `if` statement.

It is used when a second condition should be checked **only if the first condition is True**.

---

## 📌 Syntax

```python
if condition1:
    if condition2:
        # Code
```

````python
# ==========================================================
# Program 4 : Nested if Statement
# ==========================================================

# Take attendance percentage
attendance = int(input("Enter Attendance (%): "))

# Check first condition
if attendance >= 75:

    fee_paid = input("Have you paid the fees? (yes/no): ")

    # Check second condition
    if fee_paid.lower() == "yes":
        print("✅ You are eligible for the exam.")
    else:
        print("❌ Please pay your fees first.")

else:
    print("❌ Attendance is below 75%.")
````

# 🔹 Comparison Operators

Comparison operators compare two values.

The result is always **True** or **False**.

| Operator | Meaning | Example |
|----------|---------|---------|
| `==` | Equal to | `5 == 5` |
| `!=` | Not Equal | `5 != 3` |
| `>` | Greater Than | `10 > 5` |
| `<` | Less Than | `4 < 8` |
| `>=` | Greater Than or Equal | `5 >= 5` |
| `<=` | Less Than or Equal | `2 <= 3` |

````python
# ==========================================================
# Program 5 : Comparison Operators
# ==========================================================

a = 15
b = 10

print("a == b :", a == b)
print("a != b :", a != b)
print("a > b  :", a > b)
print("a < b  :", a < b)
print("a >= b :", a >= b)
print("a <= b :", a <= b)
````

# 🔹 Logical Operators

Logical operators combine multiple conditions.

Python provides three logical operators.

| Operator | Meaning |
|----------|---------|
| `and` | Both conditions must be True |
| `or` | At least one condition must be True |
| `not` | Reverses the result |

````python
# ==========================================================
# Program 6 : Logical Operators
# ==========================================================

age = 20
citizen = True

# AND Operator
print("AND :", age >= 18 and citizen)

# OR Operator
print("OR  :", age >= 18 or citizen)

# NOT Operator
print("NOT :", not citizen)
````

# 🔁 for Loop

A **for loop** is used to repeat a block of code a specific number of times.

It is commonly used when the number of iterations is already known.

---

## 📌 Syntax

```python
for variable in range(start, stop, step):
    # Code
```

````python
# ==========================================================
# Program 7 : for Loop
# ==========================================================

# Print numbers from 1 to 5

for number in range(1, 6):
    print(number)
    
````

## 💡 Understanding range()

`range()` generates a sequence of numbers.

| Code | Output |
|------|--------|
| `range(5)` | 0 1 2 3 4 |
| `range(1,6)` | 1 2 3 4 5 |
| `range(2,11,2)` | 2 4 6 8 10 |

---

### 📌 Examples

```python
range(5)
```

Output

```
0 1 2 3 4
```

```python
range(1,6)
```

Output

```
1 2 3 4 5
```

````python
# ==========================================================
# Program 8 : Print Even Numbers
# ==========================================================

# Print even numbers from 2 to 20

for number in range(2, 21, 2):
    print(number)
````

# 🔄 while Loop

A **while loop** executes as long as the given condition is **True**.

It is useful when the number of iterations is **not known** in advance.

---

## 📌 Syntax

```python
while condition:
    # Code
```

````python
# ==========================================================
# Program 9 : while Loop
# ==========================================================

count = 1

while count <= 5:
    print(count)
    count += 1
````

# ⛔ break Statement

The **break** statement immediately stops the loop when a condition is met.

### 🌍 Example

Searching for a book in a library.

Once the correct book is found, you stop searching.

````python
# ==========================================================
# Program 10 : break Statement
# ==========================================================

for number in range(1, 11):

    if number == 6:
        break

    print(number)
````

# ⏭️ continue Statement

The **continue** statement skips the current iteration and moves to the next one.

### 🌍 Example

Skip absent students while taking attendance.

````python
# ==========================================================
# Program 11 : continue Statement
# ==========================================================

for number in range(1, 11):

    if number == 6:
        continue

    print(number)
````
