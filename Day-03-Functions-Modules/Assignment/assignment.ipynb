Task 1 : Create a Simple Function

```python
def welcome():
    print("Welcome to Python!")

welcome()
```

```python
#Task 2 : Function with One Parameter
def greet(name):
    print("Hello,", name)

greet("Gaurav")
```

```python
#Task 3 : Function with Two Parameters
def add(a, b):
    print("Sum =", a + b)

add(10, 20)
```

```python
#Task 4 : Return Statement
def square(num):
    return num * num

result = square(5)
print(result)
```

```python
#Task 5 : Default Parameter
def welcome(name="Guest"):
    print("Welcome", name)

welcome()
welcome("Gaurav")
```

```python
#Task 6 : Keyword Arguments
def employee(name, salary):
    print("Name :", name)
    print("Salary :", salary)

employee(salary=50000, name="Gaurav")
```

```python
#Task 7 : Local Variable
def place():
    city = "Pune"
    print(city)

place()
```

```python
#Task 8 : Global Variable
course = "Python"

def show():
    print(course)

show()
```

```python
#Task 9 : Math Module
import math

print(math.sqrt(81))
print(math.pi)
```

```python
#Task 10 : Random Module
import random

print(random.randint(1, 100))
```
