 🔹 What is a Function?

A **function** is a reusable block of code that performs a specific task.

Instead of writing the same code again and again, we write it once inside a function and call it whenever needed.

## 📌 Syntax

```python
def function_name():
    # Code

function_name()
```

---

````python
# ==========================================================
# Program 1 : Creating Your First Function
# ==========================================================

# Define a function
def greet():
    print("👋 Welcome to the AI/ML Engineer Bootcamp!")

# Call the function
greet()
````

````python
# ==========================================================
# Program 2 : Reusing a Function
# ==========================================================

def welcome():
    print("Welcome to Python!")

# Calling the same function multiple times
welcome()
welcome()
welcome()
````

# 🔹 Function Parameters

A **parameter** is a variable that receives a value when a function is called.

Parameters make functions more flexible because they can work with different values.

---

## 📌 Syntax

```python
def function_name(parameter):
    # Code

function_name(value)
```

````python
# ==========================================================
# Program 3 : Function with Parameters
# ==========================================================

# Create a function
def greet(name):
    print(f"Hello, {name}! 👋")

# Call the function
greet("Gaurav")
greet("Rahul")
greet("Priya")
````

# 🔹 Multiple Parameters

A function can accept more than one parameter.

---

## 📌 Syntax

```python
def function_name(parameter1, parameter2):
    # Code
```

````python
# ==========================================================
# Program 4 : Multiple Parameters
# ==========================================================

def student(name, marks):
    print("Student Name :", name)
    print("Marks        :", marks)

student("Gaurav", 92)
````

# 🔹 Default Parameters

A **default parameter** already has a value.

If no value is passed, Python uses the default value automatically.

---

## 📌 Syntax

```python
def greet(name="Guest"):
    print(name)
```

````python
# ==========================================================
# Program 5 : Default Parameters
# ==========================================================

def greet(name="Guest"):
    print("Welcome,", name)

# Using default value
greet()

# Passing a custom value
greet("Gaurav")
````

# 🔹 Keyword Arguments

Keyword arguments allow us to pass values using the parameter names.

The order of arguments does not matter when keywords are used.

---

## 📌 Syntax

```python
function(parameter=value)
```

````python
# ==========================================================
# Program 6 : Keyword Arguments
# ==========================================================

def employee(name, salary):
    print("Name   :", name)
    print("Salary :", salary)

employee(salary=50000, name="Gaurav")
````

# 🔹 Return Statement

The **return** statement sends a value back from a function.

Unlike `print()`, `return` allows us to store and reuse the result.

---

## 📌 Syntax

```python
def function_name():
    return value
```

````python
# ==========================================================
# Program 7 : Return Statement
# ==========================================================

# Function to add two numbers
def add(a, b):
    return a + b

# Store the returned value
result = add(10, 20)

print("Sum =", result)
````

# 🔹 Local Variables

A **local variable** is created inside a function.

It can only be used inside that function.

---
# 🔹 Global Variables

A **global variable** is declared outside all functions.

It can be accessed from anywhere in the program.

---

## 📌 Syntax

```python
variable = value

def function():
    print(variable)
```

````python
# ==========================================================
# Program 8 : Local Variable
# ==========================================================

def student():

    name = "Gaurav"     # Local Variable

    print("Student Name:", name)

student()
````

````python
# ==========================================================
# Program 9 : Global Variable
# ==========================================================

course = "Python"

def show_course():

    print("Course:", course)

show_course()

print("Learning:", course)
````

# 🔹 Built-in Functions

Python provides many ready-to-use functions.

Some common built-in functions are:

| Function | Purpose |
|----------|---------|
| `print()` | Display output |
| `input()` | Take user input |
| `len()` | Find length |
| `type()` | Check data type |
| `sum()` | Add values |
| `max()` | Largest value |
| `min()` | Smallest value |
| `round()` | Round numbers |

````python
# ==========================================================
# Program 10 : Built-in Functions
# ==========================================================

numbers = [10, 20, 30, 40]

print("Length :", len(numbers))
print("Maximum:", max(numbers))
print("Minimum:", min(numbers))
print("Sum    :", sum(numbers))
````

# 🔹 Python Modules

A **module** is a Python file containing functions, variables, or classes.

Modules help organize code and avoid repetition.

---

### Popular Modules

- `math`
- `random`
- `datetime`
- `os`

### 🌍 Real-Life Example

A toolbox contains different tools.

Similarly, a module contains useful functions that can be reused.

````python
# ==========================================================
# Program 11 : Importing Modules
# ==========================================================

import math

print("Square Root of 64 :", math.sqrt(64))
print("Value of Pi       :", math.pi)
````
